# 1_Formatting SEO_Original_Dataset

## Objectives

The objective of this Jupyter Notebook is to address missing values and validate the consistency of each column by applying logical checks and cross-referencing within the original dataset.

### Importing dataset

```python
import os
import numpy as np
import pandas as pd
import zipfile
```

```python
# ===========================================
# 1. Import Libraries and Load the Dataset
# ===========================================

# Load the raw SEO dataset
original_seo = pd.read_csv(r"data\Original SEO ISSUES Dataset RAW\SEO ISSUES 200001 - 202312v3.0.Exclcsv.csv")

# Display dataset info: column names, non-null count, data types
original_seo.info()

# Show first 10 rows of the dataset
original_seo.head(10)
```

```python
# ===========================================
# 2. Handle Missing Values
# ===========================================

# Check total missing values per column
original_seo.isnull().sum()

# Fill missing 'Shares_Offered_All_Mkts' using values from 'Shares_Offered_This_Mkt'
original_seo['Shares_Offered_All_Mkts'] = original_seo['Shares_Offered_All_Mkts'].fillna(original_seo['Shares_Offered_This_Mkt'])

# Fill missing 'Issuer_Ticker' using 'New_Issues_Ticker'
original_seo['Issuer_Ticker'] = original_seo['Issuer_Ticker'].fillna(original_seo['New_Issues_Ticker'])

# Replace missing values in 'Price_Date_Diff' with 0
original_seo['Price_Date_Diff'] = original_seo['Price_Date_Diff'].fillna(0)

# Preview the cleaned dataset
original_seo.head()
```

```python
# ===========================================
# 3. Column Validation and Cleaning
# ===========================================

# Find rows where Security_Type mismatches between the two markets
original_seo[original_seo['Security_Type_This_Mkt'] != original_seo['Security_Type_All_Mkts']]

# Decision: Drop 'Security_Type_All_Mkts' since it has fewer distinct values
```

```python
# ===========================================
# 4. Clean Invalid 'Offer_Price_Available' Values
# ===========================================

# Define a list of invalid indicators
invalid_vals = ['invalid', 'n/a', 'na', 'no']

# Replace 'Offer_Price_USD' with NaN if the corresponding 'Offer_Price_Available' is invalid
original_seo.loc[
    original_seo['Offer_Price_Available'].str.lower().isin(invalid_vals),
    'Offer_Price_USD'
] = np.nan

# Standardize invalid values in 'Offer_Price_Available' column as "missing"
original_seo['Offer_Price_Available'] = original_seo['Offer_Price_Available'].apply(
    lambda x: 'missing' if str(x).strip().lower() in invalid_vals else x
)
```

```python
# ===========================================
# 5. Drop Unnecessary Columns
# ===========================================
original_seo.drop(columns=[
    "Free_Float_at_IPO", "Secondary_Amt_Offered_All_Mkts", "Secondary_Amt_Offered_This_Mkt",
    "Offer_Price_Available", "IPO_flag", "original_IPO_flag", "BB_Deal_Number",
    "Not_For_Profit_Flag", "PE_Backed_IPO_Issue_Flag", "Target_Market",
    "Security_Type_All_Mkts", "Deal_PermID"
], inplace=True)

# Confirm column drops
original_seo.info()
```

```python
# ===========================================
# 6. Check for Ticker Mismatches
# ===========================================
# Compare cleaned 'Issuer_Ticker' with 'New_Issues_Ticker' (case insensitive)
mismatched_rows = original_seo[
    original_seo['Issuer_Ticker'].str.strip().str.upper() != original_seo['New_Issues_Ticker'].str.strip().str.upper()
]

# Preview mismatches
mismatched_rows.head(5)
```

```python
# ===========================================
# 7. Export Unique CUSIPs to Text File
# ===========================================

# Define folder and filename for output
output_text = r'txt files'
text_file = 'unique_issuer_cusip9.txt'
output_full_path = os.path.join(output_text, text_file)

# Save unique CUSIP9 values (after stripping and type conversion) to .txt file
original_seo['Issuer_CUSIP9'].dropna().astype(str).str.strip().unique().tofile(
    output_full_path, sep='\n', format='%s'
)
```

```python
# ===========================================
# 9. Load External CUSIP9-Ticker Mapping File
# ===========================================

# Load zipped file containing CUSIP9-to-ticker mappings
file_path = r'data\Updated SEO ISSUES Dataset\Wharton Datasets\cusip9_ticker_data.zip'

with zipfile.ZipFile(file_path) as z:
    with z.open('dtljsdzeletkaqco.csv') as f:
        cusip9_ticker = pd.read_csv(f)

# Keep only unique cusip-ticker pairs
unique_cusip9_rows = cusip9_ticker[['cusip', 'tic']].drop_duplicates(subset=['cusip'])
```

```python
# ===========================================
# 10. Merge CUSIP9-Ticker Data with Main Dataset
# ===========================================

# Merge external CUSIP data to the main dataset on 'Issuer_CUSIP9'
merged_df = pd.merge(unique_cusip9_rows, original_seo, 
                     how='right', left_on='cusip', right_on='Issuer_CUSIP9')

# Fill missing ticker values using previously cleaned Issuer_Ticker
merged_df['tic'] = merged_df['tic'].fillna(merged_df['Issuer_Ticker'])
```

```python
# ===========================================
# 11. Drop and Rename Columns
# ===========================================

# Drop redundant and low-value columns
original_seo = merged_df.drop(columns=[
    "cusip", "Issuer_Ticker", "SIC", "New_Issues_Company_ID",
    "Issuer_CUSIP6", "New_Issues_Ticker", "Issuer_Exchange"
])

# Rename the merged 'tic' column to 'Issuer_Ticker'
original_seo.rename(columns={'tic': 'Issuer_Ticker'}, inplace=True)

# Drop rows where Issuer_Ticker is still missing (~3% rows)
original_seo = original_seo.dropna(subset=['Issuer_Ticker']).reset_index(drop=True)
```

```python
# ===========================================
# 12. Set Correct Data Types
# ===========================================

# Convert numeric and date fields
original_seo['Price_Date_Diff'] = original_seo['Price_Date_Diff'].astype('int64')
original_seo['Issue_Date'] = pd.to_datetime(original_seo['Issue_Date'], format='%d/%m/%Y')

# Convert categorical fields
categorical_cols = [
    'New_Issues_Primary_Exchange', 'Issue_Type', 'VC_Backed_IPO_Issue_Flag',
    'Marketplace', 'Market_Area', 'Issuer_SDC_Industry', 'Security_Type_This_Mkt'
]
for col in categorical_cols:
    original_seo[col] = original_seo[col].astype('category')

original_seo.info()
```

```python
# ===========================================
# 13. Load External Security Ticker Data
# ===========================================

file_path = r'data\Updated SEO ISSUES Dataset\Wharton Datasets\security_data_using_issue_ticker.zip'

with zipfile.ZipFile(file_path) as z:
    with z.open('meabriv7ju9f5juy.csv') as f:
        ext_security_ticker = pd.read_csv(f)

# Drop duplicate tickers to avoid multiple matches
ext_security_ticker = ext_security_ticker.drop_duplicates(subset='tic')
```

```python
# ===========================================
# 14. Merge Additional CUSIP Info from Security File
# ===========================================

# Merge external cusip info from ticker dataset
original_seo = pd.merge(
    original_seo,
    ext_security_ticker[['tic', 'cusip']],
    how='left',
    left_on='Issuer_Ticker',
    right_on='tic'
)

# Fill missing CUSIP9 using values from external data
original_seo['Issuer_CUSIP9'] = original_seo['Issuer_CUSIP9'].fillna(original_seo['cusip'])

# Clean up helper columns
original_seo.drop(columns=['tic', 'cusip'], inplace=True)
original_seo.info()
```

```python
# ===========================================
# 15. Final Clean-up and Export
# ===========================================

# Enforce proper dtypes again
original_seo['Price_Date_Diff'] = original_seo['Price_Date_Diff'].astype('int64')
original_seo['Issue_Date'] = pd.to_datetime(original_seo['Issue_Date'], format='%d/%m/%Y')
#original_seo['Offer_Year'] = pd.to_datetime(original_seo['Offer_Year'], errors='coerce')
original_seo['New_Issues_Primary_Exchange'] = original_seo['New_Issues_Primary_Exchange'].astype('category')
original_seo['Issue_Type'] = original_seo['Issue_Type'].astype('category')
original_seo['VC_Backed_IPO_Issue_Flag'] = original_seo['VC_Backed_IPO_Issue_Flag'].astype('category')
original_seo['Marketplace'] = original_seo['Marketplace'].astype('category')
original_seo['Market_Area'] = original_seo['Market_Area'].astype('category')
original_seo['Issuer_SDC_Industry'] = original_seo['Issuer_SDC_Industry'].astype('category')
original_seo['Security_Type_This_Mkt'] = original_seo['Security_Type_This_Mkt'].astype('category')
original_seo.info()
original_seo['Issuer_CUSIP9'] = original_seo['Issuer_CUSIP9'].astype(str)
original_seo.reset_index(drop=True, inplace=True)

# Convert CUSIP9 to string and reset index
original_seo['Issuer_CUSIP9'] = original_seo['Issuer_CUSIP9'].astype(str)
original_seo.reset_index(drop=True, inplace=True)

# Export final dataset as CSV and Parquet
output_path = r'data\Original SEO ISSUES Dataset RAW'
csv_filename = r'SEO_ISSUES_Formatted.csv'
parquet_filename = r'SEO_ISSUES_Formatted.parquet'

original_seo.to_csv(os.path.join(output_path, csv_filename), index=False)
original_seo.to_parquet(os.path.join(output_path, parquet_filename), index=False)
```
